```python
df["word_count"] = df['Name'].str.split().str.len()

plt.figure(figsize=(6,4))
sns.histplot(df["word_count"].dropna(), bins=5, kde=True)
plt.title("Distribution of Word Counts")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

```python
df["char_len"] = df["Name"].str.len()

plt.figure(figsize=(6,4))
sns.histplot(df["char_len"].dropna(), bins=5, kde=True)
plt.title("Distribution of String Lengths")
plt.xlabel("Length of Comment")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(18, 8))
ax = sns.countplot(x = df['Category'])
plt.xticks(rotation=45, ha='right')

plt.grid(axis='y', linestyle='--', alpha=0.7)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12, color='black')
```

```python
plt.figure(figsize=(20, 8))
genre_counts = df['Availability'].value_counts()

plt.pie(genre_counts, labels=genre_counts.index, autopct='%.2f', startangle=140, colors=sns.color_palette('colorblind'))
plt.title('Genre Distribution of Most Watched Anime')
plt.axis('equal')  
plt.show()
```
