```python
# Create a histogram of the 'age' column
sns.histplot(data=dft, x='age', kde=True)
plt.title('Distribution of Age on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
df = pd.DataFrame({
    "marks": [45, 67, 89, 45, 76, 88, 92, 56, 73, 67]
})

# Histogram for one column
df["marks"].hist(bins=5, edgecolor="black")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of Marks")
plt.show()
```

```python
sns.histplot(data=dft, x='fare', kde=True)
plt.title('Distribution of Age on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
for column in dft.select_dtypes(include=['number']).columns: # Select only numerical columns
    plt.figure(figsize=(3, 6)) # Create a new figure for each histogram
    plt.hist(dft[column].dropna(), bins=10, edgecolor='black') # Use dft and .dropna() to handle missing values
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
sns.histplot(data = dft, x ='Marks',kde = True)
plt.title('Distribution of Age on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
dft.hist(bins=4, figsize=(16, 6), edgecolor="black")
plt.suptitle("Histograms of All Numeric Columns")
plt.show()
```

```python
# Select only numerical columns
numeric_cols = dft.select_dtypes(include=np.number).columns

# Define sample size and random state
sample_n = 100
RANDOM_STATE = 42

# Take a sample of the numerical columns and create histograms




dft_sample = dft[numeric_cols].sample(min(len(dft), sample_n), random_state=RANDOM_STATE)
dft_sample.hist(bins=30, figsize=(12, 8))
plt.suptitle("Numeric feature distributions (sample)")
plt.tight_layout()
plt.show()
```
